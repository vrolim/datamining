```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier, export_graphviz
from sklearn.externals.six import StringIO
from IPython.display import Image
from subprocess import call
import matplotlib.pyplot as plt
import pydotplus
```

## Leitura das bases de dados

```python
train = pd.read_csv('RL_train.csv')
train.drop('Unnamed: 0', axis=1, inplace=True)
train.head()
```

```python
test = pd.read_csv('RL_test.csv')
test.drop('Unnamed: 0', axis=1, inplace=True)
test.head()
```

```python
X_train = train.drop('Alvo', axis=1)
y_train = train['Alvo']

X_test = test.drop('Alvo', axis=1)
y_test = test['Alvo']
```

```python
X_test = X_test.apply(lambda x: x.fillna(x.median()))
```

## Treinar Árvore de Decisão

```python
tree = DecisionTreeClassifier()
tree.fit(X_train, y_train)
```

```python
tree.predict(X_test)
```

```python
cols = X_train.columns
importances = tree.feature_importances_
pd.Series(data=importances, index=cols).sort_values(ascending=False).plot(kind='bar', figsize=(20,25))
plt.savefig('DF_importances.png')
```

## Regras

```python
dot_data = StringIO()
export_graphviz(tree, out_file=dot_data, feature_names=X_train.columns.to_list(), class_names=['SHOW-UP', 'NO-SHOW'],
                filled=True, rounded=True, proportion=False, max_depth=3)
```

```python
graph = pydotplus.graph_from_dot_data(dot_data.getvalue()) 
```

```python
Image(graph.create_png())
```

```python
graph.write_png('regras_arvore_3.png')
```
